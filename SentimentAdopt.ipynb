# Sentiment Analysis

```python
analyser = SentimentIntensityAnalyzer()

tweets = ['Fuck you nigger',"omg That's cool and amaizing","warui","bad","im selling",'Noooo',"stock is crashing!!","im so sad"]


negative_tweets = [
    "SUI is absolutely rekt today! Massive sell-off everywhere. #SUI #CryptoBloodbath",
    "What a disaster, SUI is dumping 20% in hours. Everyone panic selling! #SUI #CryptoCrash",
    "SUI plummeting hard, nothing is holding the price. Market chaos! #SUI #BearMarket",
    "This SUI crash is brutal, token bleeding out fast. Investors are terrified. #SUI #CryptoRuin",
    "SUI completely wiped out its gains. Total panic and sell-off happening now. #SUI #CryptoDisaster"
]
sentiments = []
for t in negative_tweets:
    sentiments.append(analyser.polarity_scores(t)['compound'])

sentiments

sum_sentiments = sum(sentiments)
len_sentiments = len(sentiments)

avg_sentiment = sum_sentiments/len_sentiments
print(avg_sentiment)

# Learning Point 

# Use List Comprehension
# sentiments = [ analyser.polarity_scores(t) for t in tweets ]
# sentiment

# analyser.polarity_scores('str') returns dict of feelings

# sum( np object )
# len( np object )

```

# Adoption of Coin Analysis

1.Adoption Proxy, that is the Binance Trading Volume 

any better ?

```python
url = "https://api.binance.com/api/v3/klines"

params = {"symbol":"SUIUSDT",
          "interval" : "1d",
          "limit":1000,
          }

data = requests.get(url,params=params).json()

df = pd.DataFrame(data, columns=["time","open","high","low","close","volume",
                                 "close_time","qav","num_trades","taker_base_vol","taker_quote_vol","ignore"])
df
```

2.Normalising Adoption Score

adoption score is based on :

latest trading volume as relative to max trading vol

```python
df['volume']=df['volume'].astype(float)

volume_window = df['volume'].iloc[-10:-1]

last_vol =df['volume'].iloc[-2] # -2 as yesterday's vol is used to avoid changes
max_vol = df['volume'].max()

adoption_score = last_vol/max_vol
adoption_score

volume_window
```
